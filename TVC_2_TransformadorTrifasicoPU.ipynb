## 🔢 Dados do Sistema

### 🔌 Transformador T1
- Potência nominal: 600 kVA  
- Tensão: 480 V / 34.5 kV  
- Conexão: Y-Delta  
- Resistência longitudinal: 1%  
- Reatância longitudinal: 10%  

### 🔌 Transformador T2
- Potência nominal: 400 kVA  
- Tensão: 23.1 kV / 127 V  
- Conexão: Delta-Y  
- Resistência longitudinal: 2%  
- Reatância longitudinal: 15%  

### ⚡ Linha de Distribuição (LD)
- Impedância:  
  $$
  Z_{LD} = 1 + j2 \ \Omega
  $$

### ℹ️ Impedância Transversal (núcleo magnético)
- Igual a **1000 vezes** a longitudinal (para resistências e reatâncias de T1 e T2)

---

## 🧾 Condição Operativa

- O transformador T2 opera com **tensão nominal**
- A carga consome **80% da potência nominal**
- Fator de potência da carga: **0.90 adiantado**

---

## 🧮 Bases de Impedância \( Zbase) – Para conversão em pu

A impedância base é dada por:

$$
Z_{\text{base}} = \frac{V_{\text{base}}^2}{S_{\text{base}}}
$$

---

### 🔷 T1 (lado de baixa tensão)

- \( Vbase1 = 480 V)  
- \( Sbase1 = 600 kVA)

$$
Z_{\text{base,T1}} = \frac{480^2}{600000} \quad [\Omega]
$$

---

### 🔷 T2 (lado de alta tensão)

- \( Vbase2 = 23100 V)  
- \( Sbase2 = 400 kVA)

$$
Z_{\text{base,T2}} = \frac{23100^2}{400000} \quad [\Omega]
$$

---

### 🔷 Linha de Distribuição (LD)

A impedância será convertida para pu usando a base de T2:

$$
Z_{LD,pu} = \frac{Z_{LD}}{Z_{\text{base,T2}}}
$$

---

## ✏️ (a) Regulação de Tensão de T1 (%)

A **regulação de tensão** é dada por:

$$
\text{Regulação} = \frac{V_{\text{nl}} - V_{\text{fl}}}{V_{\text{fl}}} \cdot 100
$$

Onde:

- \( Vnl): tensão sem carga (open circuit)  
- \( Vfl \): tensão com carga (80% da potência nominal, fp = 0.90 adiantado)  

A regulação é calculada com base no **lado de baixa tensão de T1** (480 V).  
Levar em conta a queda de tensão causada pelas impedâncias em pu.

---
## ✏️ (b) Rendimento de T2 (%)

O **rendimento (eficiência)** é calculado por:

$$
\eta = \frac{P_{\text{útil}}}{P_{\text{útil}} + P_{\text{perdas}}} \cdot 100
$$

Com:

$$
P_{\text{útil}} = 0.8 \cdot 400000 \cdot 0.9 = 288000 \, \text{W}
$$

E:

$$
P_{\text{perdas}} = P_{\text{Joule}} + P_{\text{núcleo}}
$$

$$
P_{\text{Joule}} = I_{\text{pu}}^2 \cdot R_{\text{pu}} \cdot S_{\text{base}}
$$

---

## ✏️ (c) Circuito Equivalente em pu

Construir o circuito equivalente total em pu, incluindo:

- Transformador T1 (recalcular impedâncias para base de T2)
- Linha LD (já em base de T2)
- Transformador T2

---

### ⚠️ Ajuste de Impedância para Nova Base

Caso necessário, converter impedâncias entre bases diferentes com:

$$
Z_{\text{pu,nova}} = Z_{\text{pu,antiga}} \cdot \left( \frac{S_{\text{base,nova}}}{S_{\text{base,antiga}}} \right) \cdot \left( \frac{V_{\text{base,antiga}}^2}{V_{\text{base,nova}}^2} \right)
$$

---

## ✏️ (d) Potência obtida no Ensaio a Vazio (W)

O ensaio a vazio é realizado no **lado de baixa tensão de T1 (480 V)** e considera apenas a **impedância transversal**.

### 🧮 Resistência transversal:

$$
R_m = 1000 \cdot R_1 = 1000 \cdot 0.01 \cdot Z_{\text{base,T1}}
$$

### 🧮 Potência ativa no ensaio:

$$
P_{\text{vazio}} = \frac{V^2}{R_m} = \frac{480^2}{R_m} \quad [W]
$$

---

## ✅ Conclusão

Com as fórmulas acima, é possível resolver as 4 letras da prova utilizando os conceitos de:

- Sistema por unidade (pu)
- Regulação de tensão
- Rendimento de transformador
- Conversão de base de impedância
- Análise de ensaio a vazio

```python
# ENE-079 – Conversão Eletromecânica de Energia I
# Prova TVC-3 – Transformadores, Linha de Distribuição e Análise em pu

import numpy as np

# --- Dados do sistema ---
# Transformador T1
S1 = 600e3         # VA
V1 = 480           # V (lado baixa tensão)
R1_percent = 1     # %
X1_percent = 10    # %

# Transformador T2
S2 = 400e3         # VA
V2 = 23100         # V (lado alta tensão)
R2_percent = 2     # %
X2_percent = 15    # %

# Linha de Distribuição
Z_LD = 1 + 2j      # ohms

# Impedância transversal (núcleo magnético) = 1000 vezes a longitudinal
# para resistências e reatâncias dos transformadores

# Condição Operativa
P_load = 0.8 * S2           # potência aparente da carga
pf_load = 0.90              # fator de potência adiantado (leading)
P_util = P_load * pf_load   # potência útil (ativa) em W

# --- Bases de impedância ---
Zbase_T1 = V1**2 / S1       # ohms
Zbase_T2 = V2**2 / S2       # ohms

print(f"Base de impedância T1 (lado BT): {Zbase_T1:.6f} Ω")
print(f"Base de impedância T2 (lado AT): {Zbase_T2:.6f} Ω")

# --- Impedâncias longitudinais em ohms ---
R1_ohm = R1_percent/100 * Zbase_T1
X1_ohm = X1_percent/100 * Zbase_T1
Z1_ohm = R1_ohm + 1j * X1_ohm

R2_ohm = R2_percent/100 * Zbase_T2
X2_ohm = X2_percent/100 * Zbase_T2
Z2_ohm = R2_ohm + 1j * X2_ohm

print(f"\nImpedância longitudinal T1 (Ω): {Z1_ohm:.6f}")
print(f"Impedância longitudinal T2 (Ω): {Z2_ohm:.6f}")
print(f"Impedância linha de distribuição (Ω): {Z_LD}")

# --- (a) Regulação de Tensão de T1 (%) ---
# Passo 1: converter impedâncias T1 para base T2

# Formula para conversão de impedâncias pu entre bases diferentes:
def Z_pu_nova(Z_pu_antiga, S_nova, S_antiga, V_antiga, V_nova):
    return Z_pu_antiga * (S_nova/S_antiga) * (V_antiga**2 / V_nova**2)

# Impedâncias T1 em pu na base T1
Z1_pu_T1 = complex(R1_percent/100, X1_percent/100)

# Converter Z1_pu para base T2
Z1_pu_T2 = Z_pu_nova(Z1_pu_T1, S2, S1, V1, V2)

print(f"\nImpedância longitudinal T1 em pu na base T2: {Z1_pu_T2:.6f}")

# Impedância da linha já em base T2
Z_LD_pu = Z_LD / Zbase_T2
print(f"Impedância linha de distribuição em pu na base T2: {Z_LD_pu:.6f}")

# Impedância longitudinal T2 já em base T2 (em pu)
Z2_pu_T2 = complex(R2_percent/100, X2_percent/100)

# Para regulação de tensão de T1, vamos calcular a queda de tensão
# do lado baixa tensão, então precisamos calcular a corrente de carga
# e tensão na base baixa tensão.

# Corrente nominal do transformador T1 na base baixa tensão:
I1_base = S1 / (np.sqrt(3) * V1)   # corrente nominal linha (A)
# Corrente da carga (80% da potência nominal de T2, mas na base T2)
S_load = P_load / pf_load  # S = P / pf
I_load = S_load / (np.sqrt(3) * V2)  # corrente na base alta tensão

# Vamos trabalhar em pu na base T2, mas queremos saber a regulação no lado BT do T1.

# Para simplificar o cálculo da regulação:
# Vnl = tensão sem carga = 1.0 pu
# Vfl = tensão com carga, que será:
# Vfl = Vnl - queda de tensão nas impedâncias em pu

# Carga em pu na base T2 (80% da nominal)
S_load_pu = 0.8

# Ângulo da carga (fp 0.9 adiantado):
# cosφ = 0.9 leading => φ = -acos(0.9)
phi_load = -np.arccos(pf_load)

# Corrente em pu (na base T2):
I_load_pu = S_load_pu * (np.cos(phi_load) - 1j * np.sin(phi_load))  # forma retangular

# Queda de tensão total em pu (em série) = impedância T1 + linha + impedância T2
# Mas para regulação de T1 (lado baixa tensão) consideramos só a queda na impedância de T1?
# No enunciado diz "considerar as impedâncias em pu", vou considerar as impedâncias do sistema que afetam o lado BT de T1.

# Como T1 é Y-Delta, T2 é Delta-Y, o sistema é:
# Rede --- T1 (Y-Delta) --- Linha --- T2 (Delta-Y) --- Carga
# A regulação pedida é do lado baixa tensão do T1, logo a tensão nominal do lado BT é 480 V.

# Podemos considerar a corrente e queda de tensão no lado BT do T1 (lado 480 V).

# Para cálculo da regulação, usaremos as impedâncias do lado BT em pu na base T1:
# Para isso, devemos converter as impedâncias da linha e T2 para a base T1

# Convertemos impedâncias da linha e do T2 para base T1:

Z_LD_pu_T1 = Z_pu_nova(Z_LD_pu, S1, S2, V2, V1)
Z2_pu_T1 = Z_pu_nova(Z2_pu_T2, S1, S2, V2, V1)

print(f"\nImpedância linha em pu na base T1: {Z_LD_pu_T1:.6f}")
print(f"Impedância T2 em pu na base T1: {Z2_pu_T1:.6f}")

# Impedância total vista do lado BT do T1:
Z_total_pu = Z1_pu_T1 + Z_LD_pu_T1 + Z2_pu_T1

# Queda de tensão em pu:
V_drop = I_load_pu * Z_total_pu

# Tensão com carga:
V_fl = 1 - V_drop

# Regulação (%)
regulacao = (abs(1) - abs(V_fl)) / abs(V_fl) * 100

print(f"\nRegulação de tensão de T1 (%): {regulacao:.2f}%")

# --- (b) Rendimento de T2 (%)

# Primeiro, calcular corrente em pu na base de T2:
I_pu_T2 = abs(I_load_pu)

# Resistência longitudinal de T2 em pu:
R2_pu = R2_percent / 100

# Potência de perdas Joule
P_joule = I_pu_T2**2 * R2_pu * S2

# Resistência transversal (núcleo magnético):
R_m = 1000 * R2_pu * Zbase_T2   # em ohms
# Note: enunciado diz transversal é 1000 vezes longitudinal para T1 e T2, mas no enunciado do rendimento só soma as perdas Joule + núcleo em T2
# Para núcleo em W, calculamos a potência ativa no núcleo, mas não foi dado diretamente o valor de perdas do núcleo para T2
# Como enunciado só diz núcleo magnético = 1000 vezes a longitudinal, consideramos perdas do núcleo proporcionais a esta resistência
# A corrente na baixa tensão de T1 é diferente, aqui foco em T2.

# Para calcular as perdas do núcleo:
# Precisamos da corrente nominal do transformador T2:
I_nom_T2 = S2 / (np.sqrt(3) * V2)
# Corrente em ampères:
I_T2 = I_pu_T2 * I_nom_T2

# Perdas no núcleo = I^2 * R_m
P_nucleo = I_T2**2 * R_m

# Potência total de perdas
P_perdas = P_joule + P_nucleo

# Rendimento:
eta = P_util / (P_util + P_perdas) * 100

print(f"\nRendimento de T2 (%): {eta:.2f}%")

# --- (c) Circuito equivalente em pu (total)

# Já temos impedâncias em pu na base T2:
print("\nCircuito equivalente em pu (base T2):")
print(f"T1 longitudinal: {Z1_pu_T2:.6f}")
print(f"Linha de distribuição: {Z_LD_pu:.6f}")
print(f"T2 longitudinal: {Z2_pu_T2:.6f}")

# --- (d) Potência no ensaio a vazio (W)

# Resistência transversal do T1:
R1_pu = R1_percent/100
R_m_T1 = 1000 * R1_pu * Zbase_T1

# Potência ativa no ensaio:
P_vazio = V1**2 / R_m_T1

print(f"\nPotência ativa no ensaio a vazio do T1 (W): {P_vazio:.2f}")
```
