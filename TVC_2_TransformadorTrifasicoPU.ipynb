##  Dados do Sistema

- Potncia nominal: 600 kVA

###  Transformador T2
- Potncia nominal: 400 kVA  
- Tenso: 23.1 kV / 127 V  
- Conexo: Delta-Y  
- Resistncia longitudinal: 2%  
- Reatncia longitudinal: 15%  

###  Linha de Distribuio (LD)
- Impedncia:  
  $$
  Z_{LD} = 1 + j2 \ \Omega
  $$

###  Impedncia Transversal (ncleo magntico)
- Igual a **1000 vezes** a longitudinal (para resistncias e reatncias de T1 e T2)

---

##  Condio Operativa

- O transformador T2 opera com **tenso nominal**
- A carga consome **80% da potncia nominal**
- Fator de potncia da carga: **0.90 adiantado**

---

##  Bases de Impedncia \( Zbase)  Para converso em pu

A impedncia base  dada por:

$$
Z_{\text{base}} = \frac{V_{\text{base}}^2}{S_{\text{base}}}
$$

---

###  T1 (lado de baixa tenso)

- \( Vbase1 = 480 V)  
- \( Sbase1 = 600 kVA)

$$
Z_{\text{base,T1}} = \frac{480^2}{600000} \quad [\Omega]
$$

---

###  T2 (lado de alta tenso)

- \( Vbase2 = 23100 V)  
- \( Sbase2 = 400 kVA)

$$
Z_{\text{base,T2}} = \frac{23100^2}{400000} \quad [\Omega]
$$

---

###  Linha de Distribuio (LD)

A impedncia ser convertida para pu usando a base de T2:

$$
Z_{LD,pu} = \frac{Z_{LD}}{Z_{\text{base,T2}}}
$$

---

##  (a) Regulao de Tenso de T1 (%)

A **regulao de tenso**  dada por:

$$
\text{Regulao} = \frac{V_{\text{nl}} - V_{\text{fl}}}{V_{\text{fl}}} \cdot 100
$$

Onde:

- \( Vnl): tenso sem carga (open circuit)  
- \( Vfl \): tenso com carga (80% da potncia nominal, fp = 0.90 adiantado)  

A regulao  calculada com base no **lado de baixa tenso de T1** (480 V).  
Levar em conta a queda de tenso causada pelas impedncias em pu.

---
##  (b) Rendimento de T2 (%)

O **rendimento (eficincia)**  calculado por:

$$
\eta = \frac{P_{\text{til}}}{P_{\text{til}} + P_{\text{perdas}}} \cdot 100
$$

Com:

$$
P_{\text{til}} = 0.8 \cdot 400000 \cdot 0.9 = 288000 \, \text{W}
$$

E:

$$
P_{\text{perdas}} = P_{\text{Joule}} + P_{\text{ncleo}}
$$

$$
P_{\text{Joule}} = I_{\text{pu}}^2 \cdot R_{\text{pu}} \cdot S_{\text{base}}
$$

---

##  (c) Circuito Equivalente em pu

Construir o circuito equivalente total em pu, incluindo:

- Transformador T1 (recalcular impedncias para base de T2)
- Linha LD (j em base de T2)
- Transformador T2

---

###  Ajuste de Impedncia para Nova Base

Caso necessrio, converter impedncias entre bases diferentes com:

$$
Z_{\text{pu,nova}} = Z_{\text{pu,antiga}} \cdot \left( \frac{S_{\text{base,nova}}}{S_{\text{base,antiga}}} \right) \cdot \left( \frac{V_{\text{base,antiga}}^2}{V_{\text{base,nova}}^2} \right)
$$

---

##  (d) Potncia obtida no Ensaio a Vazio (W)

O ensaio a vazio  realizado no **lado de baixa tenso de T1 (480 V)** e considera apenas a **impedncia transversal**.

###  Resistncia transversal:

$$
R_m = 1000 \cdot R_1 = 1000 \cdot 0.01 \cdot Z_{\text{base,T1}}
$$

###  Potncia ativa no ensaio:

$$
P_{\text{vazio}} = \frac{V^2}{R_m} = \frac{480^2}{R_m} \quad [W]
$$

```python
# ENE-079  Converso Eletromecnica de Energia I
# Prova TVC-3  Transformadores, Linha de Distribuio e Anlise em pu

import numpy as np

# --- Dados do sistema ---
# Transformador T1
S1 = 600e3         # VA
V1 = 480           # V (lado baixa tenso)
R1_percent = 1     # %
X1_percent = 10    # %

# Transformador T2
S2 = 400e3         # VA
V2 = 23100         # V (lado alta tenso)
R2_percent = 2     # %
X2_percent = 15    # %

# Linha de Distribuio
Z_LD = 1 + 2j      # ohms

# Impedncia transversal (ncleo magntico) = 1000 vezes a longitudinal
# para resistncias e reatncias dos transformadores

# Condio Operativa
P_load = 0.8 * S2           # potncia aparente da carga
pf_load = 0.90              # fator de potncia adiantado (leading)
P_util = P_load * pf_load   # potncia til (ativa) em W

# --- Bases de impedncia ---
Zbase_T1 = V1**2 / S1       # ohms
Zbase_T2 = V2**2 / S2       # ohms

print(f"Base de impedncia T1 (lado BT): {Zbase_T1:.6f} ")
print(f"Base de impedncia T2 (lado AT): {Zbase_T2:.6f} ")

# --- Impedncias longitudinais em ohms ---
R1_ohm = R1_percent/100 * Zbase_T1
X1_ohm = X1_percent/100 * Zbase_T1
Z1_ohm = R1_ohm + 1j * X1_ohm

R2_ohm = R2_percent/100 * Zbase_T2
X2_ohm = X2_percent/100 * Zbase_T2
Z2_ohm = R2_ohm + 1j * X2_ohm

print(f"\nImpedncia longitudinal T1 (): {Z1_ohm:.6f}")
print(f"Impedncia longitudinal T2 (): {Z2_ohm:.6f}")
print(f"Impedncia linha de distribuio (): {Z_LD}")

# --- (a) Regulao de Tenso de T1 (%) ---
# Passo 1: converter impedncias T1 para base T2

# Formula para converso de impedncias pu entre bases diferentes:
def Z_pu_nova(Z_pu_antiga, S_nova, S_antiga, V_antiga, V_nova):
    return Z_pu_antiga * (S_nova/S_antiga) * (V_antiga**2 / V_nova**2)

# Impedncias T1 em pu na base T1
Z1_pu_T1 = complex(R1_percent/100, X1_percent/100)

# Converter Z1_pu para base T2
Z1_pu_T2 = Z_pu_nova(Z1_pu_T1, S2, S1, V1, V2)

print(f"\nImpedncia longitudinal T1 em pu na base T2: {Z1_pu_T2:.6f}")

# Impedncia da linha j em base T2
Z_LD_pu = Z_LD / Zbase_T2
print(f"Impedncia linha de distribuio em pu na base T2: {Z_LD_pu:.6f}")

# Impedncia longitudinal T2 j em base T2 (em pu)
Z2_pu_T2 = complex(R2_percent/100, X2_percent/100)

# Para regulao de tenso de T1, vamos calcular a queda de tenso
# do lado baixa tenso, ento precisamos calcular a corrente de carga
# e tenso na base baixa tenso.

# Corrente nominal do transformador T1 na base baixa tenso:
I1_base = S1 / (np.sqrt(3) * V1)   # corrente nominal linha (A)
# Corrente da carga (80% da potncia nominal de T2, mas na base T2)
S_load = P_load / pf_load  # S = P / pf
I_load = S_load / (np.sqrt(3) * V2)  # corrente na base alta tenso

# Vamos trabalhar em pu na base T2, mas queremos saber a regulao no lado BT do T1.

# Para simplificar o clculo da regulao:
# Vnl = tenso sem carga = 1.0 pu
# Vfl = tenso com carga, que ser:
# Vfl = Vnl - queda de tenso nas impedncias em pu

# Carga em pu na base T2 (80% da nominal)
S_load_pu = 0.8

# ngulo da carga (fp 0.9 adiantado):
# cos = 0.9 leading =>  = -acos(0.9)
phi_load = -np.arccos(pf_load)

# Corrente em pu (na base T2):
I_load_pu = S_load_pu * (np.cos(phi_load) - 1j * np.sin(phi_load))  # forma retangular

# Queda de tenso total em pu (em srie) = impedncia T1 + linha + impedncia T2
# Mas para regulao de T1 (lado baixa tenso) consideramos s a queda na impedncia de T1?
# No enunciado diz "considerar as impedncias em pu", vou considerar as impedncias do sistema que afetam o lado BT de T1.

# Como T1  Y-Delta, T2  Delta-Y, o sistema :
# Rede --- T1 (Y-Delta) --- Linha --- T2 (Delta-Y) --- Carga
# A regulao pedida  do lado baixa tenso do T1, logo a tenso nominal do lado BT  480 V.

# Podemos considerar a corrente e queda de tenso no lado BT do T1 (lado 480 V).

# Para clculo da regulao, usaremos as impedncias do lado BT em pu na base T1:
# Para isso, devemos converter as impedncias da linha e T2 para a base T1

# Convertemos impedncias da linha e do T2 para base T1:

Z_LD_pu_T1 = Z_pu_nova(Z_LD_pu, S1, S2, V2, V1)
Z2_pu_T1 = Z_pu_nova(Z2_pu_T2, S1, S2, V2, V1)

print(f"\nImpedncia linha em pu na base T1: {Z_LD_pu_T1:.6f}")
print(f"Impedncia T2 em pu na base T1: {Z2_pu_T1:.6f}")

# Impedncia total vista do lado BT do T1:
Z_total_pu = Z1_pu_T1 + Z_LD_pu_T1 + Z2_pu_T1

# Queda de tenso em pu:
V_drop = I_load_pu * Z_total_pu

# Tenso com carga:
V_fl = 1 - V_drop

# Regulao (%)
regulacao = (abs(1) - abs(V_fl)) / abs(V_fl) * 100

print(f"\nRegulao de tenso de T1 (%): {regulacao:.2f}%")

# --- (b) Rendimento de T2 (%)

# Primeiro, calcular corrente em pu na base de T2:
I_pu_T2 = abs(I_load_pu)

# Resistncia longitudinal de T2 em pu:
R2_pu = R2_percent / 100

# Potncia de perdas Joule
P_joule = I_pu_T2**2 * R2_pu * S2

# Resistncia transversal (ncleo magntico):
R_m = 1000 * R2_pu * Zbase_T2   # em ohms
# Note: enunciado diz transversal  1000 vezes longitudinal para T1 e T2, mas no enunciado do rendimento s soma as perdas Joule + ncleo em T2
# Para ncleo em W, calculamos a potncia ativa no ncleo, mas no foi dado diretamente o valor de perdas do ncleo para T2
# Como enunciado s diz ncleo magntico = 1000 vezes a longitudinal, consideramos perdas do ncleo proporcionais a esta resistncia
# A corrente na baixa tenso de T1  diferente, aqui foco em T2.

# Para calcular as perdas do ncleo:
# Precisamos da corrente nominal do transformador T2:
I_nom_T2 = S2 / (np.sqrt(3) * V2)
# Corrente em ampres:
I_T2 = I_pu_T2 * I_nom_T2

# Perdas no ncleo = I^2 * R_m
P_nucleo = I_T2**2 * R_m

# Potncia total de perdas
P_perdas = P_joule + P_nucleo

# Rendimento:
eta = P_util / (P_util + P_perdas) * 100

print(f"\nRendimento de T2 (%): {eta:.2f}%")

# --- (c) Circuito equivalente em pu (total)

# J temos impedncias em pu na base T2:
print("\nCircuito equivalente em pu (base T2):")
print(f"T1 longitudinal: {Z1_pu_T2:.6f}")
print(f"Linha de distribuio: {Z_LD_pu:.6f}")
print(f"T2 longitudinal: {Z2_pu_T2:.6f}")

# --- (d) Potncia no ensaio a vazio (W)

# Resistncia transversal do T1:
R1_pu = R1_percent/100
R_m_T1 = 1000 * R1_pu * Zbase_T1

# Potncia ativa no ensaio:
P_vazio = V1**2 / R_m_T1

print(f"\nPotncia ativa no ensaio a vazio do T1 (W): {P_vazio:.2f}")
```
